# Ejercicio 2: Transformar frases
Crea un código que tome la frase inicial entregada y la corrija de tal forma que quede igual a la frase final entregada. Donde:
* frase inicial es "hola, soMos el mejor cursó de de Data-Science de Coding_Dojo. yahoo!!!"
* frase final es "hola, somos el mejor curso de Data Science de Coding Dojo"

```python
frase = "hola, soMos el mejor cursó de de Data-Science de Coding_Dojo. yahoo!!!"
frase_final = "hola, somos el mejor curso de Data Science de Coding Dojo"
tildes = {'á':'a', 'é':'e', 'í':'i', 'ó':'o', 'ú':'u'}
```

```python
import string
nueva_frase = []

# Iteramos las palabras
for word in frase.split(): # frase.split() convierte cada palabra en un elemento de una lista
    aux = ''

    # Iteramos letra por letra las palabras
    for letter in word:
        # Si encontramos un punto, detenemos el ciclo de letras
        if letter == '.':
            break
        # Reemplazamos las tildes en las letras exceptuando la coma
        elif letter == ',' or letter not in string.punctuation:
            if letter in tildes.keys():
                aux += tildes[letter] # Reemplazamos las letras con tilde
            else:
                aux += letter
        # Reemplazamos guiones o guiones bajos por espacios
        elif letter == '_' or letter == '-':
            # Ponemos en mayuscula las letras que le siguen a - y _
            if aux[-1] == '-' or aux[-1] == '_':
                aux += letter.upper()
            else:
                aux += ' '

    auxiliar = ''

    for i in range(len(aux)):
        # Mantenemos las mayúsculas donde deben estar
        if i == 0 or aux[i-1] == ' ':
            auxiliar += aux[i]
        else:
            auxiliar += aux[i].lower()


    # Añadimos la palabra a la lista nueva_frase
    if not nueva_frase or auxiliar != nueva_frase[-1]:
        nueva_frase.append(auxiliar)

    # Si encontramos un punto, detenemos el ciclo de letras
    if '.' in word:
        break

frase = " ".join(nueva_frase)

# imprimimos el resultado
print(frase)

# Corroboramos que sean iguales
print(frase == frase_final)
```
